```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# **1. Attrition count**

```python
sns.countplot(x='Attrition', data=df)
plt.show()
```

# **2. Attrition by Department**

```python
sns.countplot(x='Department', hue='Attrition', data=df)
plt.show()
```

# **3. Age vs Attrition**

```python
sns.histplot(data=df, x='Age', hue='Attrition', kde=True)
plt.show()
```

# **4. Salary Impact**

```python
sns.boxplot(x='Attrition', y='MonthlyIncome', color = 'red',data=df)
plt.show()
```

### **INSIGHTS :**
*  Employees with low salary are leaving more
*  Research & Developement department has higher attrition.
*  Younger employees tend to leave more
